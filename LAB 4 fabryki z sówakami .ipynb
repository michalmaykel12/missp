# Zadanie 1
# Celem tej części ćwiczenia jest obliczenie maksymalnego zysku dla okien aluminiowych i drewnianych. Działania odbyły się na podstawie tabeli przekazanej przez prowadzącego na laboratorium



```
from pulp import *
```

# Tworzenie zmiennej „prob”, aby zawierała dane problemu.

```
prob = LpProblem("Zyski_z_okien",LpMaximize)  ## Maximize w celu zmaksymalizowania wyniku.
```

# zmienne decyzyjnych o nazwie okna alumioniowe i drewniane

```
x1 = LpVariable("Okna aluminiowe",0,None,LpInteger)
x2 = LpVariable("Okna drewniane",0,None,LpInteger)
```

# Dodanie ograniczeń 

```
prob += 3000.0*x1 + 5000.0*x2, "Zysk maksymalny"
prob += x1 <= 4.0, "Fabryka 1"
prob += 2.0*x2 <= 12.0, "Fabryka2"
prob += 3.0*x1 + 2.0*x2 <= 18.0, "Fabryka3"
```

```
prob.writeLP("Zyski_z_okien")
```

# Rozwiązanie problemu:

```
prob.solve()
```

# wynik rozwiązania 

```
print("Status:", LpStatus[prob.status])
for v in prob.variables():
    print(v.name, "=", v.varValue)
```

# zysk maksymalny 

```
print("Zysk maksymalny = ", value(prob.objective))
```

```

from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
import pandas as pd

style = {'description_width': 'initial'}

cena_okien_aluminiowych_slider = FloatSlider(min = 0, max = 10000, value = 3000, description="cena_okien_aluminiowych_slider", style = style)
cena_okien_drewnianych_slider = FloatSlider(min = 0, max = 10000, value = 5000, description="cena_okien_drewnianych_slider", style = style)
Fabryka1_czas_aluminiowe_slider = FloatSlider(min = 0, max = 10, value = 1, description="Fabryka1_czas_aluminiowe_slider", style = style)
Fabryka1_czas_drewniane_slider = FloatSlider(min = 0, max = 10, value = 0, description="Fabryka1_czas_drewniane_slider", style = style)
Fabryka1_czas_tyg_slider = FloatSlider(min = 0, max = 52, value = 4, description="Fabryka1_czas_tyg_slider", style = style)
Fabryka2_czas_aluminiowe_slider = FloatSlider(min = 0, max = 10, value = 0, description="Fabryka2_czas_aluminiowe_slider", style = style)
Fabryka2_czas_drewniane_slider = FloatSlider(min = 0, max = 10, value = 2, description="Fabryka2_czas_drewniane_slider", style = style)
Fabryka2_czas_tyg_slider = FloatSlider(min = 0, max = 52, value = 12, description="Fabryka2_czas_tyg_slider", style = style)
Fabryka3_czas_aluminiowe_slider = FloatSlider(min = 0, max = 10, value = 3, description="Fabryka2_czas_aluminiowe_slider", style = style)
Fabryka3_czas_drewniane_slider = FloatSlider(min = 0, max = 10, value = 2, description="Fabryka2_czas_drewniane_slider", style = style)
Fabryka3_czas_tyg_slider = FloatSlider(min = 0, max = 52, value = 18, description="Fabryka2_czas_tyg_slider", style = style)


from pulp import *
def dane(cena_okien_aluminiowych = 3000,
    cena_okien_drewnianych = 5000,
    Fabryka1_czas_aluminiowe = 1,
    Fabryka1_czas_drewniane = 0,
    Fabryka1_czas_tyg = 4,
    Fabryka2_czas_aluminiowe = 0,
    Fabryka2_czas_drewniane = 2,
    Fabryka2_czas_tyg = 12,
    Fabryka3_czas_aluminiowe = 3,
    Fabryka3_czas_drewniane = 2,
    Fabryka3_czas_tyg = 18):

    prob = LpProblem("Zyski z okien",LpMaximize)

    x1 = LpVariable("Okna aluminiowe",0,None, LpInteger)
    x2 = LpVariable("Okna drewniane",0,None, LpInteger)

    prob += cena_okien_aluminiowych*x1 + cena_okien_drewnianych*x2, "Maksymalny Zysk"
    prob += Fabryka1_czas_aluminiowe*x1 + Fabryka1_czas_drewniane*x2 <= Fabryka1_czas_tyg, "Fabryka 1"
    prob += Fabryka2_czas_aluminiowe*x1 + Fabryka2_czas_drewniane*x2 <= Fabryka2_czas_tyg, "Fabryka 2"
    prob += Fabryka3_czas_aluminiowe*x1 + Fabryka3_czas_drewniane*x2 <= Fabryka3_czas_tyg, "Fabryka 3"

    prob.writeLP("Okna_zyski")

    prob.solve()

    print("Status:", LpStatus[prob.status])

    for v in prob.variables():
        print(v.name, "=", v.varValue)

    print("Maksymalny zysk = ", value(prob.objective))

    shadows = [{'name':name, 'shadow price': c.pi, "slack": c.slack} for name, c in prob.constraints.items()]
    print("*"*10+"Shadow Prices"+"*"*10)
    print(pd.DataFrame(shadows))

interact(dane,
    cena_okien_aluminiowych = cena_okien_aluminiowych_slider,
    cena_okien_drewnianych = cena_okien_drewnianych_slider,
    Fabryka1_czas_aluminiowe = Fabryka1_czas_aluminiowe_slider,
    Fabryka1_czas_drewniane = Fabryka1_czas_drewniane_slider,
    Fabryka1_czas_tyg = Fabryka1_czas_tyg_slider,
    Fabryka2_czas_aluminiowe = Fabryka2_czas_aluminiowe_slider,
    Fabryka2_czas_drewniane = Fabryka2_czas_drewniane_slider,
    Fabryka2_czas_tyg = Fabryka2_czas_tyg_slider,
    Fabryka3_czas_aluminiowe = Fabryka3_czas_aluminiowe_slider,
    Fabryka3_czas_drewniane = Fabryka3_czas_drewniane_slider,
    Fabryka3_czas_tyg = Fabryka3_czas_tyg_slider)

```
